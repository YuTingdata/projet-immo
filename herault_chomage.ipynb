```python
df_csp = pd.read_csv('csp-herault.csv', delimiter=';')

df_csp.head()
```

```python
filtered_df = df_csp[df_csp['Nom Officiel Commune'] == 'Le Caylar']
filtered_df.head()
```

```python
df_csp['population'] = df_csp['population'].round().astype(int)
df_csp
```

```python
df_grouped = df_csp.groupby('Catégorie socioprofesionnelle (CSP)').size()
df_grouped
```

CSP :

Agriculteurs exploitants
Artisans, commerçant et chefs d'entreprise
Cadres et professions intellectuelles supérieures
Professions intermédiaires
Employés
Ouvriers
Retraités
Autres personnes sans activité professionnelle = chomeurs

```python

# 1. 首先對 `CODEGEO` 進行 groupby，並加總 `population`
df_grouped = df_csp.groupby(['CODGEO', 'Catégorie socioprofesionnelle (CSP)'])['population'].sum().reset_index()

df_grouped
```

```python
actif_categories = [
    'Agriculteurs exploitants', 
    'Artisans, Comm., Chefs entr', 
    'Cadres, Prof. intel. sup.', 
    'Prof. intermédiaire', 
    'Employés', 
    'Ouvriers'
]

# 2. 計算勞動力人口（只包括勞動中的人群）
# 篩選出所有屬於勞動力群體的職業類別
df_active = df_grouped[df_grouped['Catégorie socioprofesionnelle (CSP)'].isin(actif_categories)]
```

```python
# 3. 計算失業人口（無業群體）：
# 假設 "Autres personnes sans activité professionnelle" 即為失業者
df_unemployed = df_grouped[df_grouped['Catégorie socioprofesionnelle (CSP)'] == 'Autres']
```

```python
# 4. 計算每個 CODEGEO 地區的失業率
# 先計算勞動力人口總數
df_active_total = df_active.groupby('CODGEO')['population'].sum().reset_index()
df_active_total = df_active_total.rename(columns={'population': 'active_population'})
```

```python

# 再計算失業人口
df_unemployed_total = df_unemployed.groupby('CODGEO')['population'].sum().reset_index()
df_unemployed_total = df_unemployed_total.rename(columns={'population': 'unemployed_population'})
```

```python

# 5. 合併勞動力人口和失業人口數據
df_final = pd.merge(df_active_total, df_unemployed_total, on='CODGEO', how='left')
```

```python
# 6. 計算失業率
df_final['unemployment_rate'] = (df_final['unemployed_population'] / df_final['active_population']) * 100
```

```python

# 顯示結果
print(df_final[['CODGEO', 'unemployment_rate']])
```

```python
df_final['unemployment_rate'] = df_final['unemployment_rate'].round(2)
```

```python
df_final.rename(columns={'CODGEO': 'code_commune'}, inplace=True)
```
