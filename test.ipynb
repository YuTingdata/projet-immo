```python
url0 = "https://www.data.gouv.fr/fr/datasets/r/1dc47de2-b30a-4497-b9ef-448d054b5a4f" # 2024 (seulement Q1)
url1 = "https://www.data.gouv.fr/fr/datasets/r/dd516f7a-91bb-4cad-a63c-4b55cd457f4c" # 2023
url2 = "https://www.data.gouv.fr/fr/datasets/r/91fd9f09-fe6f-4ab2-9803-84a626e1faa6" # 2022
url3 = "https://www.data.gouv.fr/fr/datasets/r/7b3d20b3-77c3-4cbd-a53d-dd7cf6c1db6c" # 2021
url4 = "https://www.data.gouv.fr/fr/datasets/r/289ada73-d850-48aa-ab95-2273643f7d46" # 2020
url5 = "https://www.data.gouv.fr/fr/datasets/r/6199e1e5-6495-424a-9ddc-67e9eb06ddd4" # 2019
```

```python
# Load data from URLs
def load_data(url):
    response = requests.get(url)
    response.raise_for_status()  # Raise an exception for HTTP errors

    with zipfile.ZipFile(BytesIO(response.content)) as z:
        csv_filename = z.namelist()[0]
        with z.open(csv_filename) as csv_file:
            df = pd.read_csv(csv_file, sep='|', low_memory=False)
            df.dropna(axis=1, how='all', inplace=True)
            df = df.dropna(subset=['Code postal'])
            df['Code postal'] = df['Code postal'].astype(str)
            code_mask = df['Code postal'].isin(codes_postaux_metropole)
            df =df[code_mask]
            df_metropoleMP['year'] = df_metropoleMP['Date mutation'].dt.year
            return df
    
df0 = load_data(url0)
df1 = load_data(url1)
df2 = load_data(url2)
df3 = load_data(url3)
df4 = load_data(url4)
df5 = load_data(url5)
```
